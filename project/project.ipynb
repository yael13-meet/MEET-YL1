The dataset is about the connection between smoking and death from lung cancer. I took the dataset from http://lib.stat.cmu.edu/DASL/Datafiles/cigcancerdat.html and it is taken from different countries in the United States.

```python
import csv
f = open('cancer.csv', 'rb')
reader = csv.reader(f)
x = []
y = []

for row in reader:
    row = row[0].split()
    #print row
    a=row[1]
    #print a
    x.append(a)
    b=row[3]
    y.append(b)
#print x
#print y
x=map(float,x)
y=map(float,y)
#print x    
#print y
f.close()
   
    
```

```python
import pylab



fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('The connection between cigarettes and cancer')
ax.set_xlabel('Number of cigarettes smoked (hds per capita)')
ax.set_ylabel('Deathes per 100K population from lung cancer')
ax.scatter(x, y)
fig.savefig("oyster_with_titles.png")
```

From the graph, we can learn that the chances to die from lung cancer is getting higher when you smoke more often. It is not exact, but you can see that there is a big difference betweem the chances for people who barely smoke and people who smoke a lot.  

```python
f = open('cancer.csv', 'rb')
reader = csv.reader(f)
p=[]
z=[]

for row in reader:
    row = row[0].split()
#     print row
    a=row[0]
    #print a
    p.append(a)
    b=row[1]
    z.append(b)
z=map(float,z)
#print p    
#print z
f.close()
```

```python
import pylab

fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_title('The amount of smokers in different countries in the United States')
ax.set_xlabel('Name of the country')
ax.set_ylabel('Number of cigarettes smoked (hds per capita)')
ax.set_xticklabels(p)
ax.bar(range(44), z, width = 0.75)
fig.savefig("oyster_with_titles.png")
```
